### Let's create the IV for this analysis - Output gap (1 and 2) ----- Fix

```python
pd.concat?
```

```python
Lags = 4 #quarters 
int_lags=  pd.concat([econ_data.interest_rate_quarter, econ_data.interest_rate_quarter.shift(Lags), econ_data.interest_rate_quarter.shift(Lags*2)],axis=1)
#Extracting only the first lag
int_lag1 = int_lags.iloc[:,[1]]
int_lag1
#Second lag
int_lag2 = int_lags.iloc[:,[2]]
int_lag2
```

### Let's first run and OLS with these lag affect as a benchmark:

```python
simple_ols = smf.ols("interest_rate_quarter ~ us_gdp_seas + gdp_seas + unemp_seas + int_lag2+ int_lag1 + year", data=econ_data)
ols_fit = simple_ols.fit()
ols_predict = ols_fit.predict()
print ols_fit.summary()
```

### We are going to use the first and the second lag effect of output gap instead of the last (t=0) output gap (--> correlated with e)
