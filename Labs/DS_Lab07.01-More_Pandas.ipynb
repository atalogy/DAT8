```python
import pandas as pd
import numpy as np

```

```python
unames = ['user_id',  'age','gender', 'occupation', 'zip']
users = pd.read_table('../data/ml-data/users.dat',
                      sep='|', header=None, names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('../data/ml-data/ratings.dat',
                        sep='\t', header=None, names=rnames)

mnames = ["movie id","movie title","release date","video release date",
        "IMDb URL","unknown","Action","Adventure","Animation",
        "Children's","Comedy","Crime","Documentary","Drama","Fantasy",
        "Film-Noir","Horror","Musical","Mystery","Romance","Sci-Fi",
        "Thriller","War","Western"]

movies = pd.read_table('../data/ml-data/movies.dat',
                       sep='|', header=None, names=mnames)

# show how one of them looks
users.head(5) 
```

```python
def extract_year(x):
    if type(x) == str :
        return int(x[-4:]) 
    else :
        return x
```

```python
movies['release year']=movies['release date'].apply(extract_year)
movies['release year'].head()
```

```python
def extract_non_string_mask(x):
    if type(x) == str :
        return False
    else :
        return True
```

```python
movies['release date'][movies['release date'].apply(extract_non_string_mask)].head()
```

```python
date_cut_off=movies['release year'] >= 1998
movies[date_cut_off].tail()
```

```python
user_ratings=ratings.merge(users,how='left',on='user_id')
```

```python
user_ratings.head()
```

```python
user_rating_movies=user_ratings.merge(movies,how='left',left_on='movie_id',right_on='movie id')
```
